```python
#N-Gram
```

```python
from nltk.util import ngrams
from collections import Counter
```

```python
def extract_gram(data,num):
    data= " ".join(data)
    n_grams=ngrams(nltk.word_tokenize(data),num)
    return [ " ".join(grams) for grams in n_grams]
```

```python
#print("1-gram: ",extract_gram(df['Message without punctuation'],2))
```

```python
def Count(n):
    n_gram=extract_gram(df['Message without punctuation'],n)
    print(Counter(n_gram))
```

```python
#UniGram
Count(1)
```

```python
#BiGram
Count(2)
```

```python
#Trigram
Count(3)
```

```python
x_train,x_test,y_train,y_test=train_test_split(features,df['Category0|1'],test_size=0.2)
print(x_train.shape)
print(x_test.shape)
```
